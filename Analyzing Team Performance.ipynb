```python
import pandas as pd
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

The frist group of metrics we are going to analyze are the Expected Goals (xG)family, this family consists of Expected Goals(xG), Expected Goals Against (xGA), Expected Goals Difference(xGD). We are going to look at the relationship they have with final league position.

# Data Collection Process

# 2022/2023 Season

```python
xg2023 = pd.read_csv('XG 2023.csv')
print(xg2023.head())
```

 Looking at the correlation matrix for xG , xGA ,xGD and position for the 2022/2023 Season

```python
correlation_matrix = xg2023[['xG #', 'xGA #', 'xGD #', 'Position']].corr()

heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm_r', vmin=0, vmax=1)
plt.title('2022/2023 Expected Correlation Matrix')
plt.savefig('2022_2023_eCorrelation_Matrix.png')
plt.show()
```

```python
correlation_matrix = xg2023[['GS #', 'GA #', 'GD #', 'Position']].corr()

heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm_r', vmin=0, vmax=1)
plt.title('2022/2023 Actual Correlation Matrix')
plt.savefig('2022_2023_aCorrelation_Matrix.png')
plt.show()
```
